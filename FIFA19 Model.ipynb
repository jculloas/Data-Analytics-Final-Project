```python
# Eliminate Repetitive and Non Value Columns
fifa19.drop(["sofifa_id"], axis = 1, inplace = True)
fifa19.drop(["long_name"], axis = 1, inplace = True)
fifa19.drop(["player_url"], axis = 1, inplace = True)
fifa19.drop(["dob"], axis = 1, inplace = True)
fifa19.drop(["wage_eur"], axis = 1, inplace = True)
fifa19.drop(["team_position"], axis = 1, inplace = True)
fifa19.drop(["body_type"], axis = 1, inplace = True)
fifa19.drop(["real_face"], axis = 1, inplace = True)
fifa19.drop(["release_clause_eur"], axis = 1, inplace = True)
fifa19.drop(["player_tags"], axis = 1, inplace = True)
fifa19.drop(["loaned_from"], axis = 1, inplace = True)
fifa19.drop(["joined"], axis = 1, inplace = True)
fifa19.drop(["contract_valid_until"], axis = 1, inplace = True)
fifa19.drop(["nation_position"], axis = 1, inplace = True)
fifa19.drop(["player_traits"], axis = 1, inplace = True)
fifa19.drop(["team_jersey_number"], axis = 1, inplace = True)
fifa19.drop(["nation_jersey_number"], axis = 1, inplace = True)
fifa19.drop(["work_rate"], axis = 1, inplace = True)
fifa19.drop(['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 
             'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb'], axis=1, inplace=True)
fifa19.head()
```

```python
fifa19.info()
```

```python
# Checking Missing Values
fifa19.isna().any().any()
```

```python
# Identifying Columns with Missing Values
fifa19.columns[fifa19.isnull().any()] 
```

```python
# Fixing Missing Values for GK
fifa19['gk_diving'].fillna(0, inplace=True)
fifa19['gk_handling'].fillna(0, inplace=True)
fifa19['gk_kicking'].fillna(0, inplace=True)
fifa19['gk_reflexes'].fillna(0, inplace=True)
fifa19['gk_speed'].fillna(0, inplace=True)
fifa19['gk_positioning'].fillna(0, inplace=True)
```

```python
# Fixing Missing Values Rating Variables - using average of the lowest scores in all 6 rating variables 
fifa19['pace'].fillna(22, inplace=True)
fifa19['shooting'].fillna(22, inplace=True)
fifa19['passing'].fillna(22, inplace=True)
fifa19['dribbling'].fillna(22, inplace=True)
fifa19['defending'].fillna(22, inplace=True)
fifa19['physic'].fillna(22, inplace=True)
```

```python
# Eliminate other rows with missing value
# using dropna() function   
fifa19.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
# Checking Missing Values
fifa19.isna().any().any()
```

```python
fifa19.info()
```

```python
# Replacing the Preferred Foot Variable observations
fifa19['preferred_foot'] = fifa19['preferred_foot'].replace(["Right"],1)
fifa19['preferred_foot'] = fifa19['preferred_foot'].replace(["Left"],2)
```

```python
# Change the Position Names
fifa19['player_positions'] = fifa19['player_positions'].replace(['GK'],1)
fifa19['player_positions'] = fifa19['player_positions'].replace(['RB', "LB", "CB"],2)
fifa19['player_positions'] = fifa19['player_positions'].replace(['RWB', "LWB"],3)
fifa19['player_positions'] = fifa19['player_positions'].replace(['CDM', "LM", "RM", "CM"],4)
fifa19['player_positions'] = fifa19['player_positions'].replace(['CAM', "LW", "RW"],5)
fifa19['player_positions'] = fifa19['player_positions'].replace(['CF', "ST"],6)
```

```python
fifa19.info()
```

```python
#Top 30 Players present in the game
fifa19[['short_name', 'overall', "nationality", "player_positions", 'club', 'value_eur']].head(30) 
```

```python
fifa19['overall'] = fifa19['overall'].astype(float)
```

```python
# Discretize the Overall column
fifa19['tier']=pd.cut(fifa19['overall'],5,labels=[5,4,3,2,1])
pd.value_counts(fifa19.tier)
```

```python
fifa19['tier'] = fifa19['tier'].astype(int)
fifa19.head()
```

```python
# Analysing Attacking Variables for Tier 1 Players
attack = fifa19.filter(["tier",'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing', "attacking_volleys"], axis=1)
attack.drop(attack.index[attack['tier'] == 2], inplace = True)
attack.drop(attack.index[attack['tier'] == 3], inplace = True)
attack.drop(attack.index[attack['tier'] == 4], inplace = True)
attack.drop(attack.index[attack['tier'] == 5], inplace = True)
attack.drop(attack.index[attack['tier'] == 6], inplace = True)
attack.drop(["tier"], axis = 1, inplace = True)
attack = attack.rename(columns = {"attacking_crossing":"Crossing"})
attack = attack.rename(columns = {"attacking_finishing":"Finishing"})
attack = attack.rename(columns = {"attacking_heading_accuracy":"Heading"})
attack = attack.rename(columns = {"attacking_short_passing":"S Passing"})
attack = attack.rename(columns = {"attacking_volleys":"Volleys"})

value = attack.mean()
std = attack.std()

colors=["red", "green", "blue", "purple", "yellow"]
plt.axhline(y=100, zorder=0)
plt.bar(range(len(attack.columns)), value+np.abs(attack.values.min()), bottom=attack.values.min(), 
        yerr=std, align='center', alpha=0.5, color=colors)

plt.xticks(range(len(attack.columns)), attack.columns)
plt.ylabel('Attacking Skills')
plt.title('Tier 1 Average Attacking Skills')
plt.show()
```
